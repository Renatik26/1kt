```python
%%file test_user.py

import pytest, json
import requests
import pprint

def test_get_user():
  url = "https://petstore.swagger.io/v2/user/string"

  actual = requests.get(url)


  assert actual.status_code == 200
  assert actual.json()["username"] == "string"

def test_post_user():
  url = "https://petstore.swagger.io/v2/user"

  myuser = {
      "id": 1111,
      "username": "Azizov",
      "firstName": "string",
      "lastName": "string",
      "email": "string",
      "password": "string",
      "phone": "string",
      "userStatus": 1
  }
  actual = requests.post(url, json = myuser)

  assert actual.status_code == 200

def test_put_user():
  url = "https://petstore.swagger.io/v2/user/azizov"

  mynewuser = {
      "id": 1111,
      "username": "Azizov",
      "firstName": "Renat",
      "lastName": "string",
      "email": "string",
      "password": "string",
      "phone": "string",
      "userStatus": 1
  }
  actual = requests.put(url, json = mynewuser)
  assert actual.status_code == 200

  geturl = "https://petstore.swagger.io/v2/user/azizov"
  actual = requests.get(geturl)
  assert actual.json()["firstName"] == "Renat"


def test_delete_user():
  url = "https://petstore.swagger.io/v2/user/azizov"
  myuser = {
      "id": 1111,
      "username": "Azizov",
      "firstName": "Renat",
      "lastName": "string",
      "email": "string",
      "password": "string",
      "phone": "string",
      "userStatus": 1
  }
  username = myuser["username"]
  response = requests.delete(url)
  assert response.status_code == 200

```
